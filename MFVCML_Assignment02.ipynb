## 1. Import packages for plotting graphs and manipulation data

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 2. define my function: $f(x) = x^2+5x+2$

```python
def myFunction(x):
    f = x**2+5*x+2
    return f
```

## 3. define the derivative of my function: $f'(x) = 2x+5$

```python
def myDerivativeFunction(x):
    Df = 2*x+5
    return Df
```

## 4. define the first-order Taylor approximation of my function $\hat{f}(x) = f(z)+f'(z)(x-z)$

```python
def myTaylorAppox(z):
    Ta = myFunction(z) + myDerivativeFunction(z)*(x-z)
    return Ta
```

## 5. define the domain of the function: $x = [-30 : 0.1 : 30]$

```python
x = np.arange(-100, 100, 0.1)
```

## 6. compute the graph

```python
f = myFunction(x)
Df = myDerivativeFunction(x)
Ta1 = myTaylorAppox(10)
Ta2 = myTaylorAppox(-20)
Ta3 = myTaylorAppox(30)
```

## 7. plot the graphs for the function and its derivative

```python
plt.figure(1)
plt.plot(x, f, 'b', label="function")
plt.plot(x, Df, 'r', label="derivative")
plt.plot(x, Ta1, 'g', label="taylor approx at x=1")
plt.plot(x, Ta2, 'c', label="taylor approx at x=3")
plt.plot(x, Ta3, 'm', label="taylor approx at x=5")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```
